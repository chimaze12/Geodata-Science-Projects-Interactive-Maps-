The dataset is then geocoded and necessary adjustments made:

```python
df1 = df.copy()
```

```python
from typing import Type
for i, row in df1.iterrows():
  try:
    info= geocode(str(row['States Affected']), provider='nominatim')
    df1.loc[int(i), 'Latitude'] = info['geometry'].loc[0].y
    df1.loc[int(i), 'Longitude'] = info['geometry'].loc[0].x
  except TypeError:
    Print('Nan')

```

```python
import folium
```

```python
new_dataset= pd.read_csv('New_df.csv')
```

```python
new_dataset.head()
```

```python
del new_dataset['Unnamed: 0']
```

```python
new_dataset.head()
```

Using folium, the map is created:

```python
m =folium.Map(tiles= 'OpenStreetmap', zoom_start= 2)

for index, row in new_dataset.iterrows():
  folium.Marker([row['Latitude'], row['Longitude']], popup= 'State: '+row['States Affected']+ '<br>' '<br>' 'Number of Death Cases as at Dec 2021: ' + str(row['Number of Deaths'])).add_to(m)

m.save('Interactive Map on Covid Cases in Nigeria.html')
```

```python
m
```

Notice how some points on the maps do not fall in their appropriate positions on the map: Niger, Delta, and Borno states have their markers outside of their boundaries. As this is clearly a case of wrong coordinates, we then try to adjust the geometry (latitude and longitude values) of these three areas on the map from the dataset.

One way to do this is manually search the OpenStreetMap API for a more suitable location detail for these areas and replace their initial values in the dataset:



```python
new_dataset.iloc[9,5:]= [5.5859, 5.4382]
new_dataset.iloc[25,5:]= [11.864, 13.195]
new_dataset.iloc[29,5:]= [9.662, 6.520,]

```

```python
new_dataset
```

Now, we re-run the mapping code to plot the now-adjusted dataset for a more precise interactive map, providing a pop-up for the required information.

```python
m =folium.Map(tiles= 'OpenStreetmap', zoom_start= 2)

for index, row in new_dataset.iterrows():
  folium.Marker([row['Latitude'], row['Longitude']], popup= 'State: '+row['States Affected']+ '<br>' '<br>' 'Number of Death Cases as at Dec 2021 : ' + str(row['Number of Deaths'])).add_to(m)

m.save('Adjusted Interactive Map on Covid Cases in Nigeria.html')
```
